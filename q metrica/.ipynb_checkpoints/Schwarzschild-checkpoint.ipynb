```python
%matplotlib inline
from scipy.integrate import odeint 
import numpy as np 
import matplotlib.pyplot as plt
```

# Schwarzschild

EL potencial efectivo y la ecuación diferencial a resolver es:


$$V(r) = \frac{1}{2}\epsilon -\epsilon\frac{GM}{r}+\frac{L^{2}}{2r^{2}}-\frac{GML^{2}}{r^{3}}$$

$$\left(\frac{dr}{d\lambda}\right)^2+2V(r)=E^2$$


Tomando $G=1$ y $L=r^2 \frac{d\varphi}{d\lambda}$ se obtiene la siguiente ecuación diferencial\\



$$\left(\frac{dr}{d\varphi}\right)^2+2\frac{r^4}{L^2}V(r)=\frac{r^4}{L^2}E^2$$


con $u=1/r$ y derivando una segunda vez, obtengo la ecuación diferencial que quiero resolver,



$$\frac{d^2 u}{d\varphi^2}=\frac{\epsilon m}{L^2} - u + 3 m u^2$$

Donde el término $ 3m u^2$ es el término extra al caso newtoniano.


## Potencial

```python
def V_s(r,e,l,gamma):
    return 0.5*e - (e/r) + (l**2)/(2*r**2)-gamma*(l**2)/(r**3)

r = np.linspace(0.01,20,1000)
for i in range(1,6):
    plt.plot(r,V_s(r,1,i,0),"r--")
    plt.grid(True)
    plt.ylim(0,0.8)
```

# Solución de la ecuación diferencial

Iniciemos intentando solucionar la ecuación sin recuperar unidades, es decir, utilizando la ecuación 

$$\frac{d^2 u}{d\varphi^2}=\frac{\epsilon m}{L^2} - u + 3 m u^2$$

nuevamente, para resolverla se reescribe $$\frac{du}{d\varphi}=v\ \ \ \ \ \ \ \text{y} \ \ \ \ \ \ \ \ \ \frac{dv}{d\varphi}=\frac{\epsilon m}{L^2} - u + 3 m u^2$$


```python
# 1) ecuación diferencial:
def dudphi(Y,t):
    u,v = Y
    new = e * m /(l**2)
    rel = 3.*m*u
    dYdt = [v,new - u + rel]
    return dYdt

# 2) rango 
phi = np.linspace(0. ,10*np.pi,100000)

# 3) constantes
e =1.
l = 1
m=0.1

# 4) Condiciones iniciales
u0 = 1./9.         
v0 = 0.1                  # para u' 
Y0 = [u0 ,v0]

# 5) La solución!
Y = odeint(dudphi,Y0,phi)


# Grafica
x = np.cos(phi)/Y[:,0]
y = np.sin(phi)/Y[:,0]
plt.plot(x,y,"r--")
plt.xlabel=('x (AU)')
plt.ylabel=('y (AU)')
plt.grid()
#plt.ylim(-600,800)


plt.plot(2*m* np.cos(phi),2*m* np.sin(phi),"b-")
plt.show()

```

```python
from scipy import integrate
def dudphi(v,t=0):
    return np.array([v[1],3*v[0]**2-v[0]+1/(l**2)])

l = 15.
phi = np.linspace(0.01,1*np.pi,100)

phi_p_0=0.6
r_0=np.sqrt(l/phi_p_0)
print r_0
Y0 = np.array([1/r_0,0.2])

Y = integrate.odeint(dudphi,Y0,phi)



ax = plt.subplot(111, projection = "polar")
ax.plot(phi,1/Y[:,0],"r-",label = "Angulo",linewidth = 2)
#ax.plot(phi,2*np.ones(len(phi)),"b--")
```

Asumimos q = 1, luego se hace extensible para cualquier q. POr lo tanto la ecuación diferencial que queremos resolver está dad por 

$$\frac{du}{d\varphi}=\left(\frac{(1-mu)^{3}}{l}\right)\left[\frac{E^{2}}{(1-2mu)^{5}}-\frac{l^{2}u^{2}}{(1-2mu)^{2}}-\frac{\epsilon}{(1-2mu)^{3}}\right]$$


$$\frac{du}{d\varphi}=\left(\frac{(1-mu)^{3}}{l(1-2mu)^{2}}\right)\left[\frac{E^{2}}{(1-2mu)^{3}}-l^{2}u^{2}-\frac{\epsilon}{(1-2mu)}\right]$$

```python

def v(r,e,l):
    return 0.5*e - (e/r) + (l**2)/(2*r**2)-0*(l**2)/(r**3)

def fq(u,phi):
    E = 50; l = 5
    return (u/l)*np.sqrt((E**2 - 2*v(u,0,l)))

ro = 0.1*np.ones(1000)
a = 0
b = 6*np.pi

phi = np.linspace(a,b,1000)

r = odeint(fq,ro,phi)

ax = plt.subplot(111, projection = "polar")
ax.plot(phi,r,"r-")
ax.plot(phi,4*np.ones(len(phi)),"b--")
ax.plot(phi,4*np.ones(len(phi)),"b--")
```
